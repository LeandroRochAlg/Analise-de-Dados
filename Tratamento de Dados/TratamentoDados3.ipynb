```python
import pandas as pd
```

```python
df = pd.read_csv("Dados/Desemprego no brasil por trimestre.csv", delimiter=';')
df
```

```python
df_test = df.melt()
df_test
```

```python
df_test = df_test.drop(index=0)
df_test
```

```python
df_test['Meses'] = df_test['variable'].str.split(' ', expand=True)[0]
df_test
```

```python
df_test['Ano'] = df_test['variable'].str.split(' ', expand=True)[1]
df_test
```

```python
df_test['Mes'] = df_test['Meses'].str.split('-', expand=True)[2]
df_test
```

```python
df_test['Taxa'] = df_test['value'].astype(float)
df_test
```

```python
df_test = df_test.drop(columns=['variable', 'Meses', 'value'])
df_test
```

```python
df_test.to_csv("Dados Tratados/Desemprego-TRATADO.csv", index=False)
```
