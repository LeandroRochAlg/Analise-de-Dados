```python
import pandas as pd
```

```python
df = pd.read_csv("../Dados/CMO-Historical-Data-Monthly.xlsx - Monthly Prices.csv")
df
```

```python
# Obtendo o cabeçalho original e a primeira linha
original_header = df.columns
first_row = df.iloc[0]

# Concatenando o cabeçalho com a primeira linha
new_header = [f'{original_header[i]} {first_row[i]}' for i in range(len(original_header))]

# Atualizando o DataFrame para usar o novo cabeçalho
df.columns = new_header
df = df[1:]  # Removendo a primeira linha que foi usada no novo cabeçalho

# Mostrando o resultado
df
```

```python
df['Date'] = pd.to_datetime(df['Date Currency'], format='%YM%m')
df = df.drop(columns=['Date Currency'])
df
```

```python
dfData = df['Date']
df = df.drop(columns=['Date'])

# Converter as colunas para float
df = df.apply(lambda x: x.str.replace(',', '.'))
```

```python
df['Date'] = dfData
df
```

```python
df.to_csv('../Dados Tratados/CMO-Historical-TRATADO.csv', index=False)
```
