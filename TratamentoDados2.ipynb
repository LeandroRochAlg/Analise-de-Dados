```python
import pandas as pd
```

```python
df = pd.read_csv("V_EXPORTACAO_E IMPORTACAO_GERAL_2015-01_2024-12_DT20240608.csv", delimiter=';')
df
```

```python
df = df.applymap(lambda x: x.replace('\r', '') if isinstance(x, str) else x)
df
```

```python
df_exportacao = df[df['Fluxo'].str.strip() == 'Exportação']
df_exportacao
```

```python
df_exportacao.to_csv("Exportação-TRATADO.csv", index=False)
```

```python
df_import= df[df['Fluxo'].str.strip() == 'Importação']
df_import.to_csv("Importação-TRATADO.csv", index=False)
df_import
```
